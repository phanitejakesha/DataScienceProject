Seeing the table I'm thinking if it is possible to predict SRType column from Agency and Outcome values. The idea is to map Agency to SRType. The service with highest count for given Agency has more probability than other and that will be our prediction. To make things more sophisticated we can use outcome values too.

```python
#Creating nested dictionary whose keys are Agency and SRType and value will number of times that SRType occured for given agency
#Might take several minutes to create this dictionary
Agency_SRType_dic = {}
for index, row in df_drop.iterrows():
    if row['Agency'] in Agency_SRType_dic:
        if row['SRType'] in Agency_SRType_dic[row['Agency']]:
            Agency_SRType_dic[row['Agency']][row['SRType']] += 1
        else:
            Agency_SRType_dic[row['Agency']][row['SRType']] = 1
    else:
        Agency_SRType_dic[row['Agency']] = {}
        Agency_SRType_dic[row['Agency']][row['SRType']] = 1
```

```python
#Uncomment below line to check full dictionary
#print(Agency_SRType_dic)
Agency_SRType_map = {}
#The above data structure contains SRType value for given Agency
for key in Agency_SRType_dic:
    count = 0
    res = None
    for key2 in Agency_SRType_dic[key]:
        if count < Agency_SRType_dic[key][key2]:
            count = Agency_SRType_dic[key][key2]
            res = key2
    Agency_SRType_map[key] = res
```

```python
#This is output of above tedious code. It is mapping of most probable SRType for given Agency
print(Agency_SRType_map)
```

```python
#Let us see if we can map zipcodes and neighbourhood to fill missing values
zip_neighbor_dic = {}
clash = 0
for index, rows in clean_df.iterrows():
    if rows['ZipCode'] in zip_neighbor_dic:
        if zip_neighbor_dic[rows['ZipCode']] != rows['Neighborhood']:
            clash += 1
    else:
        zip_neighbor_dic[ rows['ZipCode'] ] = rows['Neighborhood']
print(clash)
```

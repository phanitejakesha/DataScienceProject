```python
#All the imports for the notebook can be found here 
import pandas as pd
import os
import folium
import numpy as np
import matplotlib as mplt
import matplotlib.pyplot as plt
import geopandas
import seaborn as sns
from shapely.geometry import Point
import gmplot
```

```python
df_vacant = pd.read_csv('Vacant_Buildings.csv') # Reading the data into the dataframe df
df_vacant = df_vacant[df_vacant['BuildingAddress'].notnull()]
df_vacant.isnull().sum(axis=0)
df_vacant['PoliceDistrict']=df_vacant['PoliceDistrict'].str.lower()
```

```python
df_vacant
```

```python
df_location = pd.DataFrame(columns=['ReferenceID','Latitude', 'Longitude'])
for i in range(0,len(df_vacant)):
    location = df_vacant.iloc[i]['Location'].split(',')
    Latitude = float(location[0][1:])
    Longitude = float(location[1][:-1])
    df_location.loc[i] = [df_vacant.iloc[i]['ReferenceID'],Latitude,Longitude]
```

```python
result = pd.concat([df_vacant, df_location],axis=1, join_axes=[df_vacant.index])
result
```

```python
m = folium.Map(location=[20,0], tiles="Mapbox Bright", zoom_start=2)
for i in range(0,len(result)):
      folium.Circle(
      location=[result.iloc[i]['Longitude'], result.iloc[i]['Latitude']],
      radius=10000,
      color='crimson',
      fill=True,
      fill_color='crimson'
   ).add_to(m)
# Save it as html
m.save('mymap.html')
```

```python
result['Coordinates'] = list(zip(result.Longitude, result.Latitude))
result['Coordinates'] = result['Coordinates'].apply(Point)
gdf = geopandas.GeoDataFrame(result, geometry='Coordinates')
gdf.plot(result['Coordinates'], color='red')
plt.show()
```

```python
corr = result.corr()
fig, ax = plt.subplots(figsize=(10, 10))
colormap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, cmap=colormap, annot=True, fmt=".2f")
plt.xticks(range(len(corr.columns)), corr.columns);
plt.yticks(range(len(corr.columns)), corr.columns)
plt.show()
```

```python
df_crime = pd.read_csv('BPD_Part_1_Victim_Based_Crime_Data.csv')
df_crime =  df_crime.drop(['crimeCaseNumber', 'Total Incidents'],axis=1)
df_crime[df_crime['Latitude'].isnull()][['Location','Neighborhood','Latitude','Longitude','Location 1']]
df_crime = df_crime[df_crime['Latitude'].notnull()]
df_crime = df_crime[df_crime['Location'].notnull()]
df_crime = df_crime[df_crime['Location 1'].notnull()]
df_crime['Description'] = df_crime['Description'].fillna('No Description Provided')
df_crime['Weapon'] = df_crime['Weapon'].fillna('No Weapon')
df_crime['Inside/Outside'] = df_crime['Inside/Outside'].fillna('No Value')
df_crime['Premise'] = df_crime['Premise'].fillna('Dont know')
df_crime['Post'] = df_crime['Post'].fillna(0)
df_crime['District'] = df_crime['District'].fillna(0)
df_crime['Neighborhood'] = df_crime['Neighborhood'].fillna('Empty')
```
